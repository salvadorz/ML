```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
NFL Predictor

How to represent the data?

Eack week consists of either 15 or 16 games. We stack up all the stats
pair-wise for each week and then stack the weeks for the entire season. The
output y, if a binary decision (for now):

    
                X                    y     Comments
    ------------------------------  ----   ----------------------------
    Week 1
    1  team_a_stats, team_b_stats   1/0    1=team_a won, 0 team_b won
    ...
    16 team_x_stats, team_y_stats   1/0    1=team_x won, 0 team_y won
    
    Week 2
    1  team_a_stats, team_b_stats   1/0    1=team_a won, 0 team_b won
    ...
    16 team_x_stats, team_y_stats   1/0    1=team_x won, 0 team_y won
    
    ...
    
    Week 16
    1  team_a_stats, team_b_stats   1/0    1=team_a won, 0 team_b won
    ...
    16 team_x_stats, team_y_stats   1/0    1=team_x won, 0 team_y won
    
"""

import os
import sys

import numpy as np

from sklearn.svm import LinearSVC
from sklearn.svm import LinearSVR
from sklearn.svm import SVC
from sklearn.svm import SVR
from sklearn.svm import NuSVC
from sklearn.svm import NuSVR
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import pylab as pl

primary_stats_file_name = "../nfl_stats/2016-PrimaryStats.csv"
OFFENSE_COLUMNS = 5
DEFENSE_COLUMNS = 5
WINS_COLUMNS    = 3
TOTAL_WINS_COL  = 12
SUPER_BOWL      = 1

scaled_offense_features = ["OYds/G"]
scaled_defense_features = ["DYds/G"]

OFFENSE_SCALER  = 1
DEFENSE_SCALER  = 1

primary_stats = np.loadtxt(primary_stats_file_name, dtype="str", delimiter=",")
primary_stats = primary_stats[1:] # loose the first row
m, n = primary_stats.shape # num rows, cols

print ("Primary Stats Filename = %s" % primary_stats_file_name)
print (m, "rows")
print (n, "columns")

# -------------------------------------------------------------------------
# Create team names, feature names and team data, with team data in floats
# -------------------------------------------------------------------------
team_names    = primary_stats[0:m, 0]
team_names    = team_names[1:] # drop column heading string "Team"

feature_names = primary_stats[0,   0:n]

team_data_str = primary_stats[1:m, 1:n]

mm, nn = team_data_str.shape # num rows, cols

print ("Team Data Array:")
print (mm, "rows")
print (nn, "columns")
print ("")
team_data = np.zeros( (mm, nn) ) # create an empty array
for row in range(0, mm) :
    for col in range(0, nn) :
        # convert strings to floats
        team_data[(row, col)] = float(team_data_str[(row, col)])

print ("Team Data:")
print (team_data)

# Offense and defense feature names
offense_feature_names = feature_names[1:OFFENSE_COLUMNS+1]
defense_feature_names = feature_names[OFFENSE_COLUMNS+1:OFFENSE_COLUMNS+DEFENSE_COLUMNS+1]

# -------------------------------------------------------------------------
# Create the Offense feature matrix
# -------------------------------------------------------------------------
offense_data = np.zeros( (mm, OFFENSE_COLUMNS) ) # create an empty array
for row in range(0, mm) :
    for col in range(0, OFFENSE_COLUMNS) :
        if (offense_feature_names[col] in scaled_offense_features) :
            offense_data[(row, col)] = (team_data[(row, col)] / OFFENSE_SCALER)
        else :
            offense_data[(row, col)] = team_data[(row, col)]
     
print ("Offense data:")
print (offense_data)

# -------------------------------------------------------------------------
# Create the Defense feature matrix
# -------------------------------------------------------------------------
defense_data = np.zeros( (mm, DEFENSE_COLUMNS) ) # create an empty array
for row in range(0, mm) :
    for col in range(0, DEFENSE_COLUMNS) :
        if (defense_feature_names[col] in scaled_defense_features) :
            defense_data[(row, col)] = (team_data[(row, OFFENSE_COLUMNS+col)] / DEFENSE_SCALER)
        else :
            defense_data[(row, col)] = team_data[(row, OFFENSE_COLUMNS+col)]

print ("Defense data:")
print (defense_data)

# -------------------------------------------------------------------------
# -------------------------------------------------------------------------
#                    This section visualizes our data. 
# -------------------------------------------------------------------------
# -------------------------------------------------------------------------

# -------------------------------------------------------------------------
# Reduce the full team data set to 2 dimensions and have a look
# -------------------------------------------------------------------------
pca_team_data = PCA(n_components=2, whiten=False).fit(team_data)
pca_team_data_2d = pca_team_data.transform(team_data)

print ("pca team data shape", pca_team_data_2d.shape)
print (pca_team_data_2d)

for i in range(0, pca_team_data_2d.shape[0]) :
    superbowl = team_data[i, OFFENSE_COLUMNS+DEFENSE_COLUMNS+WINS_COLUMNS+SUPER_BOWL-1]
    if (superbowl == 1.0) :
        # highlight the 2 teams that went to the superbowl
        team_name = team_names[i]
        c1 = pl.scatter(pca_team_data_2d[i,0], pca_team_data_2d[i,1], c="b", marker="*")
        pl.annotate (team_name, (pca_team_data_2d[i,0], pca_team_data_2d[i,1]))
    else :
        c1 = pl.scatter(pca_team_data_2d[i,0], pca_team_data_2d[i,1], c="r", marker="+")

pl.title ('Team Data, PCA 2-dimensional view')
pl.show ()

# -------------------------------------------------------------------------
# Reduce the offense data set to 2 dimensions and have a look
# -------------------------------------------------------------------------
pca_offense_data = PCA(n_components=2, whiten=False).fit(offense_data)
pca_offense_data_2d = pca_offense_data.transform(offense_data)

print ("pca offense shape", pca_offense_data_2d.shape)
print (pca_offense_data_2d)

for i in range(0, pca_offense_data_2d.shape[0]) :
    superbowl = team_data[i, OFFENSE_COLUMNS+DEFENSE_COLUMNS+WINS_COLUMNS+SUPER_BOWL-1]
    ###print ("superbowl=, i=, teamname=", superbowl, i, team_names[i])
    if (superbowl == 1.0) :
        # highlight the 2 teams that went to the superbowl
        team_name = team_names[i]
        c1 = pl.scatter(pca_offense_data_2d[i,0], pca_offense_data_2d[i,1], c="b", marker="*")
        pl.annotate (team_name, (pca_offense_data_2d[i,0], pca_offense_data_2d[i,1]))
    else :
        c1 = pl.scatter(pca_offense_data_2d[i,0], pca_offense_data_2d[i,1], c="r", marker="+")

pl.title ('Offense Data, PCA 2-dimensional view')
pl.show ()

# -------------------------------------------------------------------------
# Reduce the defense data set to 2 dimensions and have a look
# -------------------------------------------------------------------------
pca_defense_data = PCA(n_components=2, whiten=False).fit(defense_data)
pca_defense_data_2d = pca_offense_data.transform(defense_data)

print ("pca defense shape", pca_defense_data_2d.shape)
print (pca_defense_data_2d)

for i in range(0, pca_defense_data_2d.shape[0]) :
    superbowl = team_data[i, OFFENSE_COLUMNS+DEFENSE_COLUMNS+WINS_COLUMNS+SUPER_BOWL-1]
    ###print ("superbowl=, i=, teamname=", superbowl, i, team_names[i])
    if (superbowl == 1.0) :
        # highlight the 2 teams that went to the superbowl
        team_name = team_names[i]
        c1 = pl.scatter(pca_defense_data_2d[i,0], pca_defense_data_2d[i,1], c="b", marker="*")
        pl.annotate (team_name, (pca_defense_data_2d[i,0], pca_defense_data_2d[i,1]))
    else :
        c1 = pl.scatter(pca_offense_data_2d[i,0], pca_defense_data_2d[i,1], c="r", marker="+")

pl.title ('Defense Data, PCA 2-dimensional view')
pl.show ()

# -------------------------------------------------------------------------
# Force calculations
# -------------------------------------------------------------------------
from math import sqrt
from math import pow
print ("Force vectors:")

# -------------------------------------------------------------------------
# Calculate the offense force values
# The greater the value, the "BETTER" the offense
# -------------------------------------------------------------------------
offense_force_value = np.zeros(mm)
max_offense_value = 0
for i in range(mm) :
    # FIXME: Add constants for field offsets
    # FIXME: linear combination OR square root of sum-of-squares?
    # Notes:
    #    Fumbles acts to reduce the offensive force value
    offense_value = sqrt( pow(offense_data[i, 0], 2) \
                        + pow(offense_data[i, 1] / OFFENSE_SCALER, 2) \
                        + pow(offense_data[i, 2], 2) \
                        - pow(offense_data[i, 3], 2) \
                        + pow(offense_data[i, 4], 2) )
    offense_force_value[i] = offense_value
    if (offense_value > max_offense_value) :
        max_offense_value = offense_value

# normalize these to 1.0
for i in range(mm) :
    offense_force_value[i] = offense_force_value[i] / max_offense_value

# -------------------------------------------------------------------------
# Plot the offense force values
# -------------------------------------------------------------------------
x_data = np.arange(0, 32, 1)
for i in range(0, offense_data.shape[0]) :
    superbowl = team_data[i, OFFENSE_COLUMNS+DEFENSE_COLUMNS+WINS_COLUMNS+SUPER_BOWL-1]
    team_name = team_names[i]
    if (superbowl == 1.0) :
        c1 = pl.scatter(x_data[i], offense_force_value[i], c="b", marker="*")
    else :
        c1 = pl.scatter(x_data[i], offense_force_value[i], c="r", marker="+")
    pl.annotate (team_name, (x_data[i], offense_force_value[i]))

pl.title ('Offense force vectors (larger is better)')
pl.xlabel('Team')
pl.show ()

# -------------------------------------------------------------------------
# Calculate the defense force values
# The greater the value, the "WORSE" the defense
# -------------------------------------------------------------------------
defense_force_value = np.zeros(mm)
max_defense_value = 0
for i in range(mm) :
    # FIXME: Add constants for field offsets
    # FIXME: linear combination OR square root of sum-of-squares?
    # Notes:
    #   Fumbles acts to reduce the defensive force value
    #   Sack acts to reduce the defensive force value
    defense_value = sqrt( pow(defense_data[i, 0], 2) \
                        + pow(defense_data[i, 1] / DEFENSE_SCALER, 2) \
                        + pow(defense_data[i, 2], 2) \
                        - pow(defense_data[i, 3], 2) \
                        - pow(defense_data[i, 4], 2) )
    defense_force_value[i] = defense_value
    if (defense_value > max_defense_value) :
        max_defense_value = defense_value
        
# normalize these to 1.0
for i in range(mm) :
    defense_force_value[i] = defense_force_value[i] / max_defense_value        

# -------------------------------------------------------------------------
# Plot the defense force values
# -------------------------------------------------------------------------
x_data = np.arange(0, 32, 1)
for i in range(0, defense_data.shape[0]) :
    superbowl = team_data[i, OFFENSE_COLUMNS+DEFENSE_COLUMNS+WINS_COLUMNS+SUPER_BOWL-1]
    team_name = team_names[i]
    if (superbowl == 1.0) :
        c1 = pl.scatter(x_data[i], defense_force_value[i], c="b", marker="*")
    else :
        c1 = pl.scatter(x_data[i], defense_force_value[i], c="r", marker="+")
    pl.annotate (team_name, (x_data[i], defense_force_value[i]))

pl.title ('Defense force vectors (larger is worse)')
pl.xlabel('Team')   
pl.show ()

# -------------------------------------------------------------------------
# Calculate the combined force values
# The greater the value, the "better" the team
# -------------------------------------------------------------------------
combined_force_value = np.zeros(mm)
for i in range(0, team_data.shape[0]) :
    combined_force_value[i] =   offense_force_value[i]   \
                              - defense_force_value[i] \
                              + (team_data[(i, TOTAL_WINS_COL)] / 16.0 ) # CONST FIXME
                                # normalize total wins, 16 games
                                # ignoring road vs. home games for now

# -------------------------------------------------------------------------
# Plot the combined force values
# -------------------------------------------------------------------------
x_data = np.arange(0, 32, 1)
for i in range(0, combined_force_value.shape[0]) :
    superbowl = team_data[i, OFFENSE_COLUMNS+DEFENSE_COLUMNS+WINS_COLUMNS+SUPER_BOWL-1]
    team_name = team_names[i]
    if (superbowl == 1.0) :
        c1 = pl.scatter(x_data[i], combined_force_value[i], c="b", marker="*")
    else :
        c1 = pl.scatter(x_data[i], combined_force_value[i], c="r", marker="+")
    pl.annotate (team_name, (x_data[i], combined_force_value[i]))

#pl.legend ([c1], ["Data Points"])
pl.title ('Combined force vectors (larger is better)')
pl.xlabel('Team')
pl.show ()

# -------------------------------------------------------------------------
# -------------------------------------------------------------------------
#               This section attempts to learn from the data 
# -------------------------------------------------------------------------
# -------------------------------------------------------------------------

# ---------------------------------------------------------------------
# Read all these CSV fields as strings
#    FIXME - add all 16 weeks, make a loop
# ---------------------------------------------------------------------
week1_file_name = "../nfl_stats/Week1-WinLoss.csv"
week1_stats = np.loadtxt(week1_file_name, dtype="str", delimiter=",")

week2_file_name = "../nfl_stats/Week2-WinLoss.csv"
week2_stats = np.loadtxt(week2_file_name, dtype="str", delimiter=",")
         
week3_file_name = "../nfl_stats/Week3-WinLoss.csv"
week3_stats = np.loadtxt(week3_file_name, dtype="str", delimiter=",")

week4_file_name = "../nfl_stats/Week4-WinLoss.csv"
week4_stats = np.loadtxt(week4_file_name, dtype="str", delimiter=",")

week15_file_name = "../nfl_stats/Week15-WinLoss.csv"
week15_stats = np.loadtxt(week15_file_name, dtype="str", delimiter=",")

week16_file_name = "../nfl_stats/Week16-WinLoss.csv"
week16_stats = np.loadtxt(week16_file_name, dtype="str", delimiter=",")

# ----------------------------------------------------------------------
# Build up the feature matrix X and expected outcome y
# by week
# ----------------------------------------------------------------------

# Create a M deep x 28 wide array full of 0's
weeks          = 6 # of training data FIXME
games_per_week = 16
depth          = games_per_week * weeks # 16 games, 4 weeks of data

# create 1 initial row with 2X number of columns of team_data
X = np.zeros( (1, 2*team_data.shape[1]) )
X_metadata = [] # an empty list to hold array of lists of: week TeamA TeamB
first_time = 1
x_index = 0 # used and incremented by fill_Xy()

# create 1 initial row
y = np.zeros( (1, 1) )

# stats = ndarray
def fill_Xy (week_str, stats):
    global   x_index
    global   X
    global   X_metadata
    global   y
    global   first_time
    
    rows, cols = stats.shape

    # Note: first row and column of stats has text/team names, x_index points
    # so we start at index=1
    
    for i in range(1, rows) :
        for j in range (1, cols) :
            if (stats[i, j] != "0") :
                
                # We have a score (ie cell was not ==0)
                
                # grow the X and y ndarrays dynamically if this is not the first time
                if (first_time == 0) :
                    #print ("grow arrays...")
                    tmp = np.zeros( (1, 2*team_data.shape[1]) )
                    X = np.concatenate((X, tmp), axis=0)
                    
                    tmp = np.zeros( (1, 1) )
                    y = np.concatenate((y, tmp), axis=0)
                    
                    x_index += 1 # increment to the new row
                    
                first_time = 0 # clear flag when we fill the first row
                                    
                scores = stats[i, j].split("-") # returns a list with the 2 scores
 
                tmp2 = team_data.shape[1]               
                #print ("i j = ", i, j, "X shape=", X.shape, "tmp2=", tmp2, "cols=", cols)
                
                # We subtract 1 from i & j, cuz team_data is numerical only, no row/columns labels
                X[x_index,    0:tmp2]   = team_data[i-1] # copy team A stats
                X[x_index, tmp2:2*tmp2] = team_data[j-1] # copy team B stats
                
                #            week      Team A name      Team B name
                tmp_list = [ week_str, str(stats[i,0]), str(stats[0,j]) ]
                X_metadata.append (tmp_list)

                if ( float(scores[0]) > float(scores[1]) ) :
                    #X[x_index, 0:14]  = np.linspace(0.0, 0.0, 14).reshape(1, 14)
                    y[x_index]        = 1 # left-side winner
                else :
                    #X[x_index, 0:14]  = np.linspace(0.0, 0.0, 14).reshape(1, 14)
                    y[x_index]        = 0 # top-side winner
    
# Week 1
weekly_stats = week1_stats
fill_Xy ("Week 1", weekly_stats)

# Week 2
weekly_stats = week2_stats
fill_Xy ("Week 2", weekly_stats)

# Week 3
weekly_stats = week3_stats
fill_Xy ("Week 3", weekly_stats)

# Week 4
weekly_stats = week4_stats
fill_Xy ("Week 4", weekly_stats)

# Week 15
weekly_stats = week15_stats
fill_Xy ("Week 15", weekly_stats)

# Week 16
weekly_stats = week16_stats
fill_Xy("Week 16", weekly_stats)

# Here are our features and expected outcomes for weeks 1-16
print ("All Training Data X shape:", X.shape)
###print (X)
print ("X, y, week and teams:")
# rows
for i in range(0, X.shape[0]) :
    # columns
    tmp = str(i) + ": "
    sys.stdout.write(tmp)
    for j in range (0, X.shape[1]) :
        tmp = str(X[i, j]) + " "
        sys.stdout.write(tmp)
    tmp = ": " + str(y[i]) + " : " + str(X_metadata[i]) + "\n"
    sys.stdout.write(tmp)

# ----------------------------------------------------------------------
# Split the dataset into training and test portions
# ----------------------------------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=115,
                                                    test_size=0.10 )

print ("X_train shape=", X_train.shape)
print ("X_test  shape=", X_test.shape)

# The SVM classifier in scikit-learn expects a 1D array for the expected outcome
y_train = y_train.ravel()
y_test = y_test.ravel()


# ----------------------------------------------------------------------
# Train the model
# ----------------------------------------------------------------------
#svmClassifier = LinearSVR()
#svmClassifier = SVC()
#svmClassifier = SVR()
#svmClassifier = NuSVC()
#svmClassifier = NuSVR()

#class LinearSVC(penalty='l2', loss='squared_hinge',  or hinge
#         dual=True, tol=0.0001, C=1.0, multi_class='ovr', 
#          fit_intercept=True, intercept_scaling=1, class_weight=None, 
#          verbose=0, random_state=None, max_iter=1000):
svmClassifier = LinearSVC(max_iter=5000)

svmClassifier.fit(X_train, y_train) 

# Set params for predict
# C = penalty param of the error term
###LinearSVC(C=1.0, dual=False, fit_intercept=True, loss='hinge', random_state=115)
LinearSVC(C=1.0, dual=False, fit_intercept=True, loss='hinge')

"""
#class sklearn.svm.SVC(C=1.0, kernel='rbf', degree=3, gamma='auto', coef0=0.0, 
#                      shrinking=True, probability=False, tol=0.001, cache_size=200, 
#                      class_weight=None, verbose=False, max_iter=-1, 
#                      decision_function_shape=None, random_state=None)

svmClassifier = SVC()

svmClassifier.fit(X_train, y_train) 

# kernel types = linear, poly, rbf (default), sigmoid
# degree of polynomial, only used by the poly kernel
# decision function shape = ovo (one vs one) ovr (one vs rest)
SVC(C=1.0, kernel='rbf', degree=3, decision_function_shape='ovr', random_state=115)
"""

"""
#class sklearn.svm.NuSVC(nu=0.5, kernel='rbf', degree=3, gamma='auto', coef0=0.0, 
#                        shrinking=True, probability=False, tol=0.001, cache_size=200, 
#                        class_weight=None, verbose=False, max_iter=-1, 
#                        decision_function_shape=None, random_state=None)

svmClassifier = NuSVC()

svmClassifier.fit(X_train, y_train) 

# kernel types = linear, poly, rbf (default), sigmoid
# degree of polynomial, only used by the poly kernel
# decision function shape = ovo (one vs one) ovr (one vs rest)
# support vector control, default 0.5
NuSVC(nu=0.5, kernel='rbf', degree=3, decision_function_shape='ovr', random_state=115)
"""

# run the test data, make the predictions
y_predicted = svmClassifier.predict (X_test)
###print predicted.shape

print ("Expected, Predicted:")
# print predicted.reshape(16,1)
for i in range(y_predicted.shape[0]):
    print (float(y_test[i]), y_predicted[i])
print ("")
 
# measure performance of the prediction/model

accuracy = metrics.accuracy_score (y_test, y_predicted)
print ("Accuracy:")
print (accuracy)
print ("")


# --------------------------------------------------------------------
# Make a prediction based on the combined force values
# --------------------------------------------------------------------
depth = X_test.shape[0]

y_closedform_predicted = np.linspace(0.0, 0.0, depth).reshape(depth, 1)

for i in range(y_closedform_predicted.shape[0]):
    y_closedform_predicted[i] = combined_force_value[i]
```

```python
import os
import sys
import numpy as np
import pylab as pl
from math import sqrt, pow
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.decomposition import PCA

primary_stats_file_name = "../nfl_stats/2016-PrimaryStats.csv"
OFFENSE_COLUMNS = 5
DEFENSE_COLUMNS = 5
WINS_COLUMNS = 3
TOTAL_WINS_COL = 12
SUPER_BOWL = 1

scaled_offense_features = ["OYds/G"]
scaled_defense_features = ["DYds/G"]

OFFENSE_SCALER = 1
DEFENSE_SCALER = 1

primary_stats = np.loadtxt(primary_stats_file_name, dtype="str", delimiter=",")
primary_stats = primary_stats[1:]  # Remove the first row

m, n = primary_stats.shape  # Number of rows, columns

print("Primary Stats Filename =", primary_stats_file_name)
print(m, "rows")
print(n, "columns")

# Create team names, feature names, and team data, with team data in floats
team_names = primary_stats[0:m, 0]
team_names = team_names[1:]

feature_names = primary_stats[0, 0:n]

team_data_str = primary_stats[1:m, 1:n]

mm, nn = team_data_str.shape  # Number of rows, columns

print("Team Data Array:")
print(mm, "rows")
print(nn, "columns")
print("")

team_data = team_data_str.astype(float)

# Offense and defense feature names
offense_feature_names = feature_names[1:OFFENSE_COLUMNS + 1]
defense_feature_names = feature_names[OFFENSE_COLUMNS + 1:OFFENSE_COLUMNS + DEFENSE_COLUMNS + 1]

# Create the Offense feature matrix
offense_data = np.copy(team_data[:, :OFFENSE_COLUMNS])

# Scale the specified offense features
for col in range(OFFENSE_COLUMNS):
    if offense_feature_names[col] in scaled_offense_features:
        offense_data[:, col] /= OFFENSE_SCALER

print("Offense data:")
print(offense_data)

# Create the Defense feature matrix
defense_data = np.copy(team_data[:, OFFENSE_COLUMNS:OFFENSE_COLUMNS + DEFENSE_COLUMNS])

# Scale the specified defense features
for col in range(DEFENSE_COLUMNS):
    if defense_feature_names[col] in scaled_defense_features:
        defense_data[:, col] /= DEFENSE_SCALER

print("Defense data:")
print(defense_data)

# Visualize the data using PCA
pca_team_data_2d = PCA(n_components=2, whiten=False).fit_transform(team_data)
pca_offense_data_2d = PCA(n_components=2, whiten=False).fit_transform(offense_data)
pca_defense_data_2d = PCA(n_components=2, whiten=False).fit_transform(defense_data)

# Plot the PCA-transformed team data
def plot_pca_data(data_2d, labels, title):
    for i in range(data_2d.shape[0]):
        superbowl = team_data[i, OFFENSE_COLUMNS + DEFENSE_COLUMNS + WINS_COLUMNS + SUPER_BOWL - 1]
        if superbowl == 1.0:
            pl.scatter(data_2d[i, 0], data_2d[i, 1], c="b", marker="*")
            pl.annotate(labels[i], (data_2d[i, 0], data_2d[i, 1]))
        else:
            pl.scatter(data_2d[i, 0], data_2d[i, 1], c="r", marker="+")
    pl.title(title)
    pl.show()

plot_pca_data(pca_team_data_2d, team_names, 'Team Data, PCA 2-dimensional view')
plot_pca_data(pca_offense_data_2d, team_names, 'Offense Data, PCA 2-dimensional view')
plot_pca_data(pca_defense_data_2d, team_names, 'Defense Data, PCA 2-dimensional view')

# Calculate force vectors
offense_force_value = np.zeros(mm)
defense_force_value = np.zeros(mm)
combined_force_value = np.zeros(mm)
max_offense_value = 0
max_defense_value = 0

for i in range(mm):
    offense_value = sqrt(pow(offense_data[i, 0], 2)
                         + pow(offense_data[i, 1]/ OFFENSE_SCALER, 2) 
                         + pow(offense_data[i, 2], 2)
                         - pow(offense_data[i, 3], 2)
                         + pow(offense_data[i, 4], 2))
    defense_value = sqrt(pow(defense_data[i, 0], 2)
                         + pow(defense_data[i, 1], 2)
                         + pow(defense_data[i, 2], 2)
                         - pow(defense_data[i, 3], 2)
                         - pow(defense_data[i, 4], 2))

    offense_force_value[i] = offense_value
    defense_force_value[i] = defense_value
    combined_force_value[i] = offense_value - defense_value + (team_data[i, TOTAL_WINS_COL] / 16.0)

# Normalize the force values
max_offense_value = np.max(offense_force_value)
max_defense_value = np.max(defense_force_value)

offense_force_value /= max_offense_value
defense_force_value /= max_defense_value

# Plot the force values
def plot_force_values(data, title):
    x_data = np.arange(len(data))
    for i in range(len(data)):
        superbowl = team_data[i, OFFENSE_COLUMNS + DEFENSE_COLUMNS + WINS_COLUMNS + SUPER_BOWL - 1]
        if superbowl == 1.0:
            pl.scatter(x_data[i], data[i], c="b", marker="*")
        else:
            pl.scatter(x_data[i], data[i], c="r", marker="+")
        pl.annotate(team_names[i], (x_data[i], data[i]))
    pl.title(title)
    pl.xlabel('Team')
    pl.show()

plot_force_values(offense_force_value, 'Offense force vectors (larger is better)')
plot_force_values(defense_force_value, 'Defense force vectors (larger is worse)')
plot_force_values(combined_force_value, 'Combined force vectors (larger is better)')

# Create a M deep x 28 wide array full of 0's
weeks          = 6 # of training data FIXME
games_per_week = 16
depth          = games_per_week * weeks # 16 games, 4 weeks of data

# Build the feature matrix X and expected outcome y by week
X = np.zeros((depth, 2 * team_data.shape[1]))
X_metadata = []
y = np.zeros((depth, 1))
x_index = 0

def fill_Xy(week_str, stats):
    global x_index
    global X
    global X_metadata
    global y

    for i in range(1, stats.shape[0]):
        for j in range(1, stats.shape[1]):
            if stats[i, j] != "0":
                if x_index != 0:
                    X = np.vstack((X, np.zeros((1, 2 * team_data.shape[1]))))
                    y = np.vstack((y, np.zeros((1, 1))))
                    x_index += 1

                scores = stats[i, j].split("-")
                team_a_stats = team_data[i-1]
                team_b_stats = team_data[j-1]
                X[x_index, :team_data.shape[1]] = team_a_stats
                X[x_index, team_data.shape[1]:] = team_b_stats
                X_metadata.append([week_str, str(stats[i, 0]), str(stats[0, j])])

                if float(scores[0]) > float(scores[1]):
                    y[x_index] = 1
                else:
                    y[x_index] = 0

                x_index += 1

# Load weekly stats and fill X, y
weekly_stats_files = [
    "../nfl_stats/Week1-WinLoss.csv",
    "../nfl_stats/Week2-WinLoss.csv",
    "../nfl_stats/Week3-WinLoss.csv",
    "../nfl_stats/Week4-WinLoss.csv",
    "../nfl_stats/Week15-WinLoss.csv",
    "../nfl_stats/Week16-WinLoss.csv"
]

for i, file_name in enumerate(weekly_stats_files):
    week_stats = np.loadtxt(file_name, dtype="str", delimiter=",")
    week_str = "Week " + str(i + 1)
    fill_Xy(week_str, week_stats)

print("All Training Data X shape:", X.shape)

# Split the dataset into training and test portions
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=115, test_size=0.10)

print("X_train shape =", X_train.shape)
print("X_test shape =", X_test.shape)

# mismatch in the shape of the expected outcome y when evaluating the model. 
# The SVM classifier in scikit-learn expects a 1D array for the expected outcome
y_train = y_train.ravel()
y_test = y_test.ravel()


# Train the model
svmClassifier = LinearSVC(max_iter=20000)
svmClassifier.fit(X_train, y_train)
LinearSVC(C=1.0, dual=False, fit_intercept=True, loss='hinge')
y_predicted = svmClassifier.predict(X_test)

# Evaluate the model
accuracy = metrics.accuracy_score(y_test, y_predicted)
precision = metrics.precision_score(y_test, y_predicted)
recall = metrics.recall_score(y_test, y_predicted)
f1_score = metrics.f1_score(y_test, y_predicted)

print("Model Evaluation:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```
